```python
import pandas as pd
import torch
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline
```

Question Answering

```python
QA_input = [{'question': 'How Large the India is?',
             'context':'India is a country in South Asia, it is the seventh-largest country and the second most populated country in the world.'},
            {'question' : 'Which group of fish does shark falls to?',
             'context':"Bony Fish: This is the largest group of fish, including familiar species like salmon, trout, cod, and tuna. They have skeletons made of bone rather than cartilage.Cartilaginous Fish: Sharks, rays, and skates belong to this group. Their skeletons are made of cartilage instead of bone.Jawless Fish: This group includes lampreys and hagfish. They lack jaws and have a unique circular mouth with rows of teeth.Freshwater Fish: These fish inhabit rivers, lakes, and streams. Examples include trout, bass, catfish, and carp.Saltwater Fish: Found in oceans and seas, saltwater fish include tuna, swordfish, cod, and various types of sharks. "}]
```

```python
# Selecting the right Pretrained Models for my QA purpose So , Im using the Huggingface page it contains lots of pretrained library models
#this model also have different configuration of architecture in terms of no of parameters, encoder layer and hidden states
model_name  = 'deepset/roberta-base-squad2'
```

```python
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
inputs0 = tokenizer(QA_input[0]['question'],QA_input[0]['context'],return_tensors = "pt")
output0 = model(**inputs0)
```

```python
inputs1 = tokenizer(QA_input[1]['question'],QA_input[1]['context'],return_tensors="pt")
output1 = model(**inputs1)
```

```python
output0
```

```python
answer_start_idx = torch.argmax(output0.start_logits)
answer_end_idx = torch.argmax(output0.end_logits)
answer_tokens  = inputs0.input_ids[0 , answer_start_idx: answer_end_idx + 1] #here im out corresponding feature from the context
answer = tokenizer.decode(answer_tokens)# it helps to convert back token ids to text form
print("ques: {}\nanswer: {}".format(QA_input[0]['question'],answer))
```

```python
answer_start_idx = torch.argmax(output1.start_logits)
answer_end_idx = torch.argmax(output1.end_logits)
answer_tokens = inputs1.input_ids[0 , answer_start_idx: answer_end_idx + 1]
answer = tokenizer.decode(answer_tokens)
print("ques: {}\nanswer: {}".format(QA_input[1]['question'],answer))
```
